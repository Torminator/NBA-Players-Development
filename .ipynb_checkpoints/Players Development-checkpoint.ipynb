```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data = pd.read_csv("Seasons_Stats.csv")
```

```python
# useless first column (just indicies)
data.drop(data.columns[[0, 21, 26]], axis=1, inplace=True)
# Between every year there is a empty row
data.dropna(how="all", inplace=True)
```

```python
data.head()
```

```python
# There are PG-SG combination but we want only the first one
data["Pos"] = data["Pos"].apply(lambda x: x.split("-")[0])
data["Player"] = data["Player"].apply(lambda x: x.split("*")[0])
```

```python
# New column to store the number of years a player played in the NBA (important since i will cut from year 1982)
data["NBAYears"] = 0.0
```

```python
# If traded during the season we have for each team a row and a total row which seemed to be alyways the first one
# if groupby
gg = data.groupby(["Player", "Year"])
table = gg.first()
```

```python
# Compute the number of years and have one position for each player (otherwise its hard to find similar players)
def setNBAYearsAndPos(x):
    pos = x["Pos"].value_counts().argmax()
    x["NBAYears"] = list(x["Age"]- x["Age"].min() + 1)
    x["Pos"] = pos
    return x

table = table.groupby(level=0).apply(lambda x: setNBAYearsAndPos(x))
```

```python
# Transform from a pivot table back to a dataframe
df = pd.DataFrame(table.to_records())
```

```python
# Only use data from 1982 onwards
pstats = df[df["Year"] > 1981]
```

```python
# Percentage of e.g. 2P/2PA are equal to nan
# Set to 0
pstats = pstats.fillna(0)
```

```python
# Set total number as per game
colidx = [28,29,31,32,34,35,37,38,39,41,42,43,44,45,46,47,48,49]
for i in colidx:
    pstats.ix[:, i] = pstats.ix[:, i]/pstats.ix[:,5]
```

```python
def meetsCondition(x, years, start=1):
    if x["Year"].count() > years and x["NBAYears"].min() == start:
        return x
```

```python
def similiarity(x, comp, cat, years):
    mat = np.array(x[cat].head(years))
    return np.sum((comp - mat)**2)
    
```

```python
def getSimiliarPlayers(name, n):
    if name in list(pstats["Player"]):
        stats_sp = pstats[pstats["Player"] == name]
        years = len(stats_sp)
        #cat = ["MP", "BPM", "USG%", "FGA", "FG%", "3PA", "3P%", 
        #  "FTA", "FT%", "TRB", "AST", "STL", "BLK", "TOV", "PTS"]
        cat = "PER"
        stats = pstats[stats_sp["Pos"].values[0] == pstats["Pos"]].groupby("Player").apply(lambda x: meetsCondition(
            x, years)).dropna(how="all")
        maximum = stats.ix[:,5:-1].max()
        stats.ix[:,5:-1] = stats.ix[:,5:-1] / maximum
        rescaled_comp = np.array(stats_sp[cat])/np.array(maximum[cat])
        sim_players = stats.groupby("Player").apply(lambda x: similiarity(x, rescaled_comp, cat, years)).sort_values(
                ascending=True).head(n).index
        return list(sim_players)
    else:
        print("Unknown Player")
```

```python
name = "Marc Gasol"
sim_players = getSimiliarPlayers(name, 20)
```

```python
sim_players
```

```python
sim_p_data = pstats.pivot_table(index=["Player", "Year"]).loc[list(sim_players)]
stats_sp = pstats[pstats["Player"] == name]
```

```python
col = "PER"

plt.figure(figsize=(16,8))
for p in sim_players[0:7]:
    plt.plot(list(sim_p_data.loc[p][col]), label=p)

plt.plot(list(stats_sp[col]), label=name, linewidth=5.0)
sim_mean = sim_p_data.groupby("NBAYears")[col].mean()
sim_std = sim_p_data.groupby("NBAYears")[col].std()
leng = len(list(stats_sp[col]))-1
plt.fill_between(list(range(leng, len(sim_std)+1)),sim_mean[leng:]-2*sim_std[leng:], sim_mean[leng:]+2*sim_std[leng:],
                 facecolor='grey',alpha=0.4)
plt.legend()
print(plt.xlim())

if col == "PER":
    plt.text(plt.xlim()[1]+0.1, 13, "Rotation", fontsize=12)
    plt.axhline(y=13, linestyle=":", color="r")

    plt.text(plt.xlim()[1]+0.1, 15, "Slightly Above", fontsize=12)
    plt.axhline(y=15, linestyle=":", color="r")

    plt.text(plt.xlim()[1]+0.1, 16.5, "3rd Option", fontsize=12)
    plt.axhline(y=16.5, linestyle=":", color="r")

    plt.text(plt.xlim()[1]+0.1, 18, "2nd Option", fontsize=12)
    plt.axhline(y=18, linestyle=":", color="r")

    plt.text(plt.xlim()[1]+0.1, 20, "Borderline All-Star", fontsize=12)
    plt.axhline(y=20, linestyle=":", color="r")

    plt.text(plt.xlim()[1]+0.1, 22.5, "All-Star", fontsize=12)
    plt.axhline(y=22.5, linestyle=":", color="r")
```
